```python
import numpy as np

import pandas as pd
pd.set_option('display.max_rows',10)
pd.set_option('display.max_columns',100)
```

```python
flux = pd.read_csv('flux/table_flux_detections_stripe82_unique.csv')
flux.head()
```

```python
crate = pd.read_csv('countrate/table_countrates_detections_stripe82_unique.csv')
crate.head()
```

```python
import numpy as np
flux.loc[flux.objid != crate.objid]
```

```python
crate.loc[flux.objid != crate.objid]
```

```python
all(flux.index == crate.index)
```

```python
cat = pd.concat([flux,crate.drop(columns=['objid','ra','dec','exposure_time'])], axis=1)
cat
```

## Flux *versus* exposure time

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [12,9]
```

```python
fig,ax = plt.subplots()

# ax.scatter(np.log(cat.exposure_time), np.log(cat.crate_full), zorder=2)
ax.loglog(cat.exposure_time, cat.crate_full, 'o', zorder=2)

ax.set_xlabel('log(exposure-time [s])')
ax.set_ylabel('log(countrates - full band [1/s])')
ax.set_title('Sensitivity according to exposure-time')
ax.grid(True)

plt.show()
```

```python
cat.describe()
```
